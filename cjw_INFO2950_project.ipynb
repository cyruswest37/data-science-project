## Data Description

The data used in this project were downloaded on 2/24/21 from a repository on Kaggle, a platform that allows users to share data sets in public repositories. As of 3/18/21, the repository can be found at the following link: https://www.kaggle.com/nicholasmarangi/job-patterns-for-minorities-and-women-usa.

According to the repository's owner, Nicholas Marangi, the data in the repository were collected from the databases of the United States Equal Employment Opportunity Commission (EEOC) as well as [data.gov](https://www.data.gov/), a website hosting open data from the United States government. The owner of the repository has released the data set into the creative commons and the source data were released publicly by the United States government.

The data contains aggregated information from EEO-1 reports collected by the EEOC from private employers in the United States with more than 100 employees as well as federal contractors with more than 50 employees. EEO-1 reports contain information about the composition of an organization's work force categorized by both sex and race/ethnic category. Aggregated data from these reports are published yearly by the EEOC. The data sets from the Kaggle repository mentioned above are compiled information from EEOC reports for the years 2011 through 2018.

### Motivation

Based on [information from the United States Equal Employment Opportunity Commission (EEOC)](https://www.eeoc.gov/statistics/employment/jobpatterns/eeo1), the data were originally collected by mandate of Title VII of the Civil Rights Act of 1964. Under this legislation, the EEOC periodically collects EEO-1 reports from employers, unions, and other labor organizations on the composition of their work force categorized by "sex" and by race/ethnic category. As such, the United States government funded the collection of the data to assist the EEOC in its mission of enforcing federal employment discrimination law.

It is unclear what the Kaggle repository owner's motivation was for compiling this data from the EEOC.

###  Composition

For the first data set (which is derived from the CountByGeneralIndustry(NAC2).csv file), each row corresponds to the employment data from a particular industry in a particular year. The unique attributes or column variables for this data set are ind_code and ind which correspond to the NAC2 industry code of the industry and the name of industry category, respectively.

For the second data set (derived from the USTotal.csv file), each row corresponds to the the employment data for all of private industry for a particular year. The only unique attribute or column variable for this data set is year, which indicates the year to which the row or observation corresponds to.

The following attributes or column variables are shared between the two data sets:
- tot:       the total number of employees in the observation
- tot_m:     the total number of male employees in the observation
- tot_f:     the total number of female employees in the observation
- tot_white: the total number of white employees in the observation
- tot_poc:   the total number people of color in the observation
- white_m:   the number of white male employees in the observation
- white_f:   the number of white female employees in the observation
- tot_black: the total number of black employees in the observation
- black_m:   the number of black male employees in the observation
- black_f:   the number of black female employees in the observation
- tot_HL:    the total number of hispanic/latino employees in the observation
- HL_m:      the number of hispanic/latino male employees in the observation
- HL_f:      the number of hispanic/latino female employees in the observation
- tot_asian: the total number of asian employees in the observation
- asian_m:   the number of asian male employees in the observation
- asian_f:   the number of asian female employees in the observation
- tot_AI:    the total number of american indian employees in the observation
- AI_m:      the number of american indian male employees in the observation
- AI_f:      the number of american indian female employees in the observation
- tot_PI:    the total number of pacific islander employees in the observation
- PI_m:      the number of pacific islander male employees in the observation
- PI_f:      the number of pacific islander female employees in the observation
- tot_bi:    the total number of biracial employees in the observation
- bi_m:      the number of biracial male employees in the observation
- bi_f:      the number of biracial female employees in the observation

## Preregistered Analyses

* Generate linear regression model for overall private employment in United States based on data from 2011 to 2015 and use to predict values for 2016 through 2018 to compare with <abbr title="observed ratio * total employment">scaled representations</abbr> of existing data from 2016 to 2018 as substitute for the full reported demographic data - will provide some sense of what expected overall private employment might have looked like in United States during Trump presidency, given changes to EEOC reporting requirements

* Perform analysis similar to the above for each industry category referenced in research question - will provide some sense of how employment may have changed during the Trump presidency in particular industries where women and/or people of color are traditionally underrepresented

(NOTE: "Data scaling" as mentioned above involves multiplying the observed demographic ratio from the reported data with the known total employment in a given category. For example: in a sample of 100 where only 10% of demographic information is reported and the observed demographic ratio is 4/10, we naively infer that 40 members of the sample occupy one group while the other 60 occupy another.)

## Data Analysis

```python
# load libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# load data
emp_by_ind = pd.read_csv('emp_by_ind1.csv')
emp_by_year = pd.read_csv('emp_by_year1.csv')
```

```python
# plot total private employment by year
plt.scatter(emp_by_year.year, emp_by_year.tot, color = "black")
plt.title("Total US Private Employment by Year")
plt.xlabel("Year")
plt.ylabel("Total Private Employment")
plt.show()

# print descriptive statistics
print("Mean US Private Employment from 2011 to 2018 ≈ {:.0f}".format(emp_by_year['tot'].mean()))
print("Standard Deviation of US Private Employment from 2011 to 2018 ≈ {:.0f}".format(emp_by_year['tot'].std()))
```

```python
# plot total employment by "sex"
plt.scatter(emp_by_year.year, emp_by_year.tot_m, color = 'blue', label = "\"male\"", alpha = .8)
plt.scatter(emp_by_year.year, emp_by_year.tot_f, color = 'red', label = "\"female\"", alpha = .8)
plt.title("Total US Private Employment by \"Sex\"")
plt.xlabel("Year")
plt.ylabel("Total Employment")
plt.legend()
plt.show()

# print descriptive statistics
print("Mean Private Employment of 'Females' in the US from 2011 to 2018 ≈ {:.0f}".format(emp_by_year['tot_f'].mean()))
print("Standard Deviation of Private Employment of 'Females' in the US from 2011 to 2018 ≈ {:.0f}".format(emp_by_year['tot_f'].std()))
```

```python
# isolate data from before Trump presidency for use as training data
pre_2016 = emp_by_year[emp_by_year['year'] < 2016]
# isolate data during Trump presidency for scaling underreported data
post_2015 = emp_by_year[emp_by_year['year'] > 2015]
```

```python
# create function to generate and plot linear regression models for "sex" in total employment data
def plot_tot_sex():
    # create and fit linear regression models
    lr1 = LinearRegression()
    lr1.fit(pre_2016[['year']], pre_2016['tot_f'])
    lr2 = LinearRegression()
    lr2.fit(pre_2016[['year']], pre_2016['tot_m'])

    # plot data
    plt.scatter(emp_by_year.year, emp_by_year.tot_m, color = 'blue', label = "\"male\"", alpha = .8)
    plt.scatter(emp_by_year.year, emp_by_year.tot_f, color = 'red', label = "\"female\"", alpha = .8)
    plt.title("Total US Private Employment by \"Sex\"")
    plt.xlabel("Year")
    plt.ylabel("Total Employment")

    # plot regression lines
    predict = lr1.predict(emp_by_year[['year']])
    plt.plot(emp_by_year.year, predict, color = 'black', label = "regression line", alpha = .8)
    predict = lr2.predict(emp_by_year[['year']])
    plt.plot(emp_by_year.year, predict, color = 'black', alpha = .8)

    # plot scaled points
    scaled = post_2015['tot'] * (post_2015['tot_f'] / post_2015['tot_m'])
    plt.plot(post_2015.year, scaled, color = 'red', label = "scaled data (\"female\")", alpha = .8)
    scaled = post_2015['tot'] * (1 - (post_2015['tot_f'] / post_2015['tot_m']))
    plt.plot(post_2015.year, scaled, color = 'blue', label = "scaled data (\"male\")", alpha = .8)

    # generate plot legend
    plt.legend()

    # render plot
    plt.show()
    
    # print model characteristics
    print("Regression Model for Private Employment of \"Females\" in the United States:")
    print("Coefficient ≈ {:.0f}".format(lr1.coef_[0]))
    print("r^2 ≈ {:.3f}".format(lr1.score(pre_2016[['year']], pre_2016['tot_f'])))
```

```python
plot_tot_sex()
```

```python
# create function to generate and plot linear regression models for race in total employment data
def plot_tot_race():
    # create and fit linear regression models
    lr1 = LinearRegression()
    lr1.fit(pre_2016[['year']], pre_2016['tot_poc'])
    lr2 = LinearRegression()
    lr2.fit(pre_2016[['year']], pre_2016['tot_white'])

    # plot data
    plt.scatter(emp_by_year.year, emp_by_year.tot_white, color = 'orangered', label = "white", alpha = .8)
    plt.scatter(emp_by_year.year, emp_by_year.tot_poc, color = 'green', label = "non-white", alpha = .8)
    plt.title("Total White & Non-White Private Employment\n in the US")
    plt.xlabel("Year")
    plt.ylabel("Total Employment")

    # plot regression lines
    predict = lr1.predict(emp_by_year[['year']])
    plt.plot(emp_by_year.year, predict, color = 'black', label = "regression line", alpha = .8)
    predict = lr2.predict(emp_by_year[['year']])
    plt.plot(emp_by_year.year, predict, color = 'black')

    # plot scaled points
    scaled = post_2015['tot'] * (post_2015['tot_poc'] / post_2015['tot_white'])
    plt.plot(post_2015.year, scaled, color = 'green', label = "scaled data", alpha = .8)
    scaled = post_2015['tot'] * (1 - (post_2015['tot_poc'] / post_2015['tot_white']))
    plt.plot(post_2015.year, scaled, color = 'orangered', label = "scaled data", alpha = .8)

    # generate plot legend
    plt.legend()

    # render plot
    plt.show()
    
    # print model characteristics
    print("Regression Model for Private Employment of \"POC\" in the United States:")
    print("Coefficient ≈ {:.0f}".format(lr1.coef_[0]))
    print("r^2 ≈ {:.3f}".format(lr1.score(pre_2016[['year']], pre_2016['tot_poc'])))
```

```python
plot_tot_race()
```

```python
# create function to generate and plot linear regression models for "sex" given industry data
def plot_sex(ind_name, ind_data):
    # isolate data from before Trump presidency for use as training data
    pre_2016 = ind_data[ind_data['year'] < 2016]
    # isolate data during Trump presidency for scaling underreported data
    post_2015 = ind_data[ind_data['year'] > 2015]
    
    # create and fit linear regression models
    lr1 = LinearRegression()
    lr1.fit(pre_2016[['year']], pre_2016['tot_f'])
    lr2 = LinearRegression()
    lr2.fit(pre_2016[['year']], pre_2016['tot_m'])

    # plot data
    plt.scatter(ind_data.year, ind_data.tot_m, color = 'blue', label = "\"male\"", alpha = .8)
    plt.scatter(ind_data.year, ind_data.tot_f, color = 'red', label = "\"female\"", alpha = .8)
    plt.title("Private US {}\n Employment by \"Sex\"".format(ind_name))
    plt.xlabel("Year")
    plt.ylabel("Total Employment")

    # plot regression lines
    predict = lr1.predict(ind_data[['year']])
    plt.plot(ind_data.year, predict, color = 'black', label = "regression line", alpha = .8)
    predict = lr2.predict(ind_data[['year']])
    plt.plot(ind_data.year, predict, color = 'black', alpha = .8)

    # plot scaled points
    scaled = post_2015['tot'] * (post_2015['tot_f'] / (post_2015['tot_m'] + post_2015['tot_f']))
    plt.plot(post_2015.year, scaled, color = 'red', label = "scaled data (\"female\")", alpha = .8)
    scaled = post_2015['tot'] * (post_2015['tot_m'] / (post_2015['tot_f'] + post_2015['tot_m']))
    plt.plot(post_2015.year, scaled, color = 'blue', label = "scaled data (\"male\")", alpha = .8)

    # generate plot legend
    plt.legend()

    # render plot
    plt.show()
    
    # print model characteristics
    print("Regression Model for Employment of \"Females\" in the United States' Private {} Industry:".format(ind_name))
    print("Coefficient ≈ {:.0f}".format(lr1.coef_[0]))
    print("r^2 ≈ {:.3f}".format(lr1.score(pre_2016[['year']], pre_2016['tot_f'])))
```

```python
plot_sex("Professional/Scientific/Technical Services", emp_by_ind[emp_by_ind.ind_code == 54])
```

```python
plot_sex("Educational Services", emp_by_ind[emp_by_ind.ind_code == 61])
```

```python
plot_sex("Health Care/Social Assistance", emp_by_ind[emp_by_ind.ind_code == 62])
```

```python
# create function to generate and plot linear regression models for race given industry data
def plot_race(ind_name, ind_data):
    # isolate data from before Trump presidency for use as training data
    pre_2016 = ind_data[ind_data['year'] < 2016]
    # isolate data during Trump presidency for scaling underreported data
    post_2015 = ind_data[ind_data['year'] > 2015]
    
    # create and fit linear regression models
    lr1 = LinearRegression()
    lr1.fit(pre_2016[['year']], pre_2016['tot_poc'])
    lr2 = LinearRegression()
    lr2.fit(pre_2016[['year']], pre_2016['tot_white'])

    # plot data
    plt.scatter(ind_data.year, ind_data.tot_white, color = 'orangered', label = "white", alpha = .8)
    plt.scatter(ind_data.year, ind_data.tot_poc, color = 'green', label = "POC", alpha = .8)
    plt.title("Private White & POC {}\n Employment in the US".format(ind_name))
    plt.xlabel("Year")
    plt.ylabel("Total Employment")

    # plot regression lines
    predict = lr1.predict(ind_data[['year']])
    plt.plot(ind_data.year, predict, color = 'black', label = "regression line", alpha = .8)
    predict = lr2.predict(ind_data[['year']])
    plt.plot(ind_data.year, predict, color = 'black')

    # plot scaled points
    scaled = post_2015['tot'] * (post_2015['tot_poc'] / post_2015['tot_white'])
    plt.plot(post_2015.year, scaled, color = 'green', label = "scaled data (\"POC\")", alpha = .8)
    scaled = post_2015['tot'] * (post_2015['tot_white'] / (post_2015['tot_white'] + post_2015['tot_poc']))
    plt.plot(post_2015.year, scaled, color = 'orangered', label = "scaled data (\"white\")", alpha = .8)

    # generate plot legend
    plt.legend()

    # render plot
    plt.show()
    
    # print model characteristics
    print("Regression Model for Employment of POC in the United States' Private {} Industry:".format(ind_name))
    print("Coefficient ≈ {:.0f}".format(lr1.coef_[0]))
    print("r^2 ≈ {:.3f}".format(lr1.score(pre_2016[['year']], pre_2016['tot_poc'])))
```

```python
plot_race("Professional/Scientific/Technical Services", emp_by_ind[emp_by_ind.ind_code == 54])
```

```python
plot_race("Educational Services", emp_by_ind[emp_by_ind.ind_code == 61])
```

```python
plot_race("Health Care/Social Assistance", emp_by_ind[emp_by_ind.ind_code == 62])
```

Contrary to expectation, the naive approach of scaling under-disclosed data from 2016 to 2018 alluded to *increased "sex" and/or race disparities in total private employment and all target industries* in the United States. These findings *may* indicate **increased "sex" and/or race discrimination in private employment during the first three years of the Trump administration**, especially in the Professional/Scientific/Technical Services, Educational Services, and Health Care/Social Assistance industries. Ultimately, this assertion cannot be proved or disproved until the Trump administration's 2016 disclosure restrictions are reversed and the [United States Equal Employment Opportunity Commission (EEOC)](https://www.eeoc.gov/) discloses full demographic data of private employment from 2016 to 2018.

## Limitations

* During exploratory data analysis it was discovered that **there is a large discrepancy in the data reported on sex and race/ethnic category between 2015 and 2016 onward**. Research revealed the following message on the [United States Equal Employment Opportunity Commission's (EEOC's) website](https://www.eeoc.gov/statistics/employment/jobpatterns/eeo1):
>"With the release of EEOC data from 2016 forward, new disclosure criteria were implemented and applied at all aggregate types to ensure the protection of identifiable information of our survey respondents and maintain EEOC's commitment to protect the data confidentiality. Participation rates and certain industry aggregates are no longer published due to disclosure limitation rules."

    This may mean data on "sex" and race/ethnic category are available from only private employers that *chose* to disclose demographic information. This fundamental change in reporting at the beginning of the Trump era limits definitive results of analysis to large private employers who were willing to disclose employment demographic information.

* **The NAC2 industry classifications from the [North American Industry Classification System (NAICS)](https://www.census.gov/naics/?58967?yearbck=2017) are broad categories**. These broad categories include a vareity of industries, each of which may significantly alter the results of analysis. For example, the category Professional, Scientific, and Technical Services includes lawyers, accountants, engineers, interior designers, architects, photographers, etc. The results of analysis cannot be generalized to singular, acute industries.

* **Data are only provided for private industry and federal contractors**. Cross comparison of employment in private and public industries is impossible and the results of analysis will not be generalizable outside of private industry.

* **The data set does not include observations from small private employers or small federal contractors.** EEO-1 reports are only collected from private employers with 100 more employees and federal contractors with 50 or more employees. The results of analysis will not be generalizable to small private employers or small federal contractors.

* **No data are available for intersex employees in private industry.** This may be a result of flawed binary representations of sex demographic data or perhaps a deliberate effort to protect the privacy of intersex individuals in the workforce. As stated on the [United States Equal Employment Opportunity Commission's (EEOC's) website](https://www.eeoc.gov/statistics/employment/jobpatterns/eeo1),
> "The confidentiality provision which governs release of these data (Section 709 (e) of Title VII of the Civil Rights Act of 1964, as amended by the Equal Employment Opportunity Act of 1972) prohibits release of individually identifiable information. Data in aggregated format for major geographic areas and by industry group for private employers (EEO-1) are available."

    This limits the ability of the results to describe how employment of intersex individuals might have been affected by the Trump presidency.
